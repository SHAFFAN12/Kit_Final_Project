# **Duration Column Cleaning**

```python
import numpy as np

def convert_to_hours(value):
    if isinstance(value, float):
        return np.nan

    if 'hours' in str(value):
        # Extract numeric value from the string
        numeric_value = ''.join(filter(str.isdigit, str(value)))

        # Check if numeric_value is not empty
        if numeric_value:
            return float(numeric_value)
        else:
            return np.nan

    elif 'months' in str(value):
        month = float(str(value).split('months')[0].strip())
        hours_per_week = float(str(value).split('at')[-1].split('hours')[0].strip())
        return month* hours_per_week  # Corrected calculation

    else:
        return np.nan

```

```python
data['Duration'] = data['Duration'].apply(convert_to_hours)
data
```

```python
# Convert 'Review' column to numeric (errors='coerce' will replace non-convertible values to NaN)
data['Duration'] = pd.to_numeric(data['Duration'], errors='coerce')

# Calculate mean and fill NaN values
mean_value = data['Duration'].mean()
data['Duration'].fillna(mean_value, inplace=True)
data['Duration'] = data['Duration'].astype(int)

data
data = data.rename(columns = ({'Duration':"duration-in-hours"}))

data
```

# **Review Cleaning**

# ***Rating Column Cleaning***

```python
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

# Calculate mean and round it to 1 decimal place
mean_value = data['Rating'].mean()
mean_value_rounded = round(mean_value, 1)  # Round to 1 decimal place

# Fill NaN values with rounded mean value
data['Rating'].fillna(mean_value_rounded, inplace=True)

data
```

```python
data['Rating-Score'] = pd.cut(data['Rating'], bins=[0, 1.5, 2.5, 3.8,4.5,5.0], labels=['very low', 'low', 'medium', 'high', 'very high'])
data
```
